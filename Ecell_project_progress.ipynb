```python
from transformers import MarianMTModel, MarianTokenizer

model_name = "Helsinki-NLP/opus-mt-hi-en"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

text = "यह एक सरकारी अधिसूचना है।"
inputs = tokenizer(text, return_tensors="pt", padding=True)
translated = model.generate(**inputs)
print(tokenizer.decode(translated[0], skip_special_tokens=True))
```

```python
cultural_dict = {
    "Dear Sir": "श्री माननीय",
    "Yours sincerely": "आपका विश्वासी",
    "Regards": "सादर",
    "Government of India": "भारत सरकार",
    "Name": "नाम",
    "Date of Birth": "जन्म तिथि"
}
def apply_cultural_tone(text, mapping):
    for eng, local in mapping.items():
        text = text.replace(eng, local)
    return text

translated = "Dear Sir, your request has been approved.\nRegards,\nGovernment of India"
rewritten = apply_cultural_tone(translated, cultural_dict)
print(rewritten)
```
